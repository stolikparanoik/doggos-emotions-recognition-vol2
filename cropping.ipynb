```python
def crop(xml, image):
    xml = minidom.parse(xml)
    image = plt.imread(image) 
    xmin = xml.getElementsByTagName('xmin')[0].firstChild.nodeValue
    xmax = xml.getElementsByTagName('xmax')[0].firstChild.nodeValue
    ymin = xml.getElementsByTagName('ymin')[0].firstChild.nodeValue
    ymax = xml.getElementsByTagName('ymax')[0].firstChild.nodeValue
    return image[int(ymin):int(ymax), int(xmin):int(xmax), :]
```

```python
#directory of file with both xml and jpg files
dirx = r"data\smiley"
for fx in os.listdir(dirx):
    for fi in os.listdir(dirx):
        if ((fx.rsplit('.',1)[0]==fi.rsplit('.',1)[0]) and fx.endswith(".xml") and fi.endswith(".jpg")):
            print(fi + " dir: " + fx)
            try:
                #change the name after last \ to the name of desired file
                plt.imsave(r'C:\Users\a.szlendak\Desktop\pg\dogos-emotions-recognition-vol2\croppedsmiley' +'\\' + fi, crop(dirx+'\\'+fx, dirx+'\\'+fi))
            except: break
            break
        else:
            continue
```
